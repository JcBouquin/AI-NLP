```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
```

```python
import io
from google.colab import files
uploaded = files.upload()
```

```python
pdf_path = "/content/MachineLearning-Lecture01.pdf"
loader = PyPDFLoader(file_path=pdf_path)
documents = loader.load()
text_splitter = CharacterTextSplitter(
chunk_size=1000, chunk_overlap=30, separator="\n"
)
docs = text_splitter.split_documents(documents=documents)
```

```python
pip install faiss-gpu
```

```python
embeddings = OpenAIEmbeddings()
vectorstore = FAISS.from_documents(docs, embeddings)
vectorstore.save_local("faiss_index_react")
```

```python
new_vectorstore = FAISS.load_local("faiss_index_react", embeddings)
```

```python
qa = RetrievalQA.from_chain_type(llm=OpenAI(), chain_type="stuff", retriever=new_vectorstore.as_retriever())
```

```python
res = qa.run("who is the instructor ? ")
print(res)
```

```python
res = qa.run("is there a dialog in this document ? ")
print(res)
```

```python
res = qa.run("do you know what is the goal of this document ?  ")
print(res)
```
